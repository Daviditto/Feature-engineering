# Binning, Discretization, Linear Models, and Trees

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
import numpy as np
import mglearn
from matplotlib import pyplot as plt
```

```python
X,y = mglearn.datasets.make_wave(n_samples=100)
```

```python
X.shape
```

```python
line = np.linspace(-3,3,1000, endpoint=False).reshape(-1,1)
```

```python
tree = DecisionTreeRegressor(min_samples_split=3).fit(X,y)
plt.plot(line, tree.predict(line), label='tree')
linear = LinearRegression().fit(X,y)
plt.plot(line, linear.predict(line),'--r', label='linear')
plt.plot(X,y,'ok')
plt.legend(loc='best');
```

```python
bins = np.linspace(-3,3,11)
print(bins)
```

```python
which_bin=np.digitize(X, bins=bins)
```

```python
X.shape
```

```python
which_bin.shape
```

```python
print(f'data point:\n{X[:10]}')
print(f'bin memebership for each point:\n{which_bin[:10]}')
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder(sparse=False)
```

```python
encoder.fit(which_bin)
```

```python
X_binned = encoder.transform(which_bin)
```

```python
X_binned.shape
```

```python
X_binned[:10]
```

```python
line_binned = encoder.transform(np.digitize(line, bins=bins))
```

```python
linear = LinearRegression().fit(X_binned, y)
plt.plot(line, linear.predict(line_binned),'g', label='linear regression binned')
reg = DecisionTreeRegressor().fit(X_binned, y)
plt.plot(line, reg.predict(line_binned),'--r', label='decision tree binned')
plt.plot(X,y, 'ok')
plt.legend(loc='best')
plt.ylabel('Regression output')
plt.xlabel('input feature');
```

# Utilizing Expert Knowledge

```python
citibike = mglearn.datasets.load_citibike()
```

```python
citibike.head()
```

```python
plt.figure(figsize=(10,3))
xticks = pd.date_range(start=citibike.index.min(), end=citibike.index.max(), freq='D')
plt.xticks(xticks, xticks.strftime('%a %m-%d'), rotation=90, ha='left')
plt.plot(citibike, linewidth=2);
```

```python
y = citibike.values
X = citibike.index.strftime('%s').astype("int").values.reshape(-1,1)
n_train=184
```

```python
def eval_on_features(features, target, regressor):
    XTrain, XTest=features[:n_train], features[n_train:]
    yTrain, yTest=target[:n_train], target[n_train:]
    regressor.fit(XTrain,yTrain)
    print(f'test set score: {regressor.score(XTest, yTest)}')
    y_pred = regressor.predict(XTest)
    y_pred_train = regressor.predict(XTrain)
    
    plt.figure(figsize=(10,3))
    
    plt.xticks(xticks, xticks.strftime('%a %m-%d'), rotation=90, ha='left')
    
    plt.plot(range(n_train), yTrain, label='train')
    plt.plot(range(n_train,len(XTest)+n_train), yTest,'-',label='test')
    plt.plot(range(n_train), y_pred_train, ':k',label='prediction train')
    plt.plot(range(n_train,len(XTest)+n_train),y_pred, '--', label='prediction test' )
    plt.legend(loc=(1.01, 0))
    plt.xlabel('Date')
    plt.ylabel('Rentals')
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = RandomForestRegressor(n_estimators=100)
plt.figure()
eval_on_features(X,y,regressor);
```

```python
X_hour = citibike.index.hour.values.reshape(-1,1)
```

```python
eval_on_features(X_hour, y, regressor)
```

```python
X_hour_week = np.hstack([citibike.index.week.values.reshape(-1,1), citibike.index.hour.values.reshape(-1,1)])
```

```python
eval_on_features(X_hour_week, y, regressor)
```

```python
eval_on_features(X_hour_week, y, LinearRegression())
```
